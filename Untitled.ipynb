```python
import numpy as np
from numba import cuda, int32
```

```python
# Tile size (tune depending on GPU: 8, 16, 32)
TILE_SIZE = 4
```

```python
@cuda.jit
def tiled_matmul(A, B, C, n):
  # Shared memory tiles
  tileA = cuda.shared.array((TILE_SIZE, TILE_SIZE), dtype=int32)
  tileB = cuda.shared.array((TILE_SIZE, TILE_SIZE), dtype=int32)

  #Thread indices
  tx=cuda.threadIdx.x
  ty=cuda.threadIdx.y
  row = cuda.blockIdx.y * TILE_SIZE + ty
  col = cuda.blockIdx.x * TILE_SIZE + tx

  temp=0

  # Loop over tiles
  for m in range((n + TILE_SIZE - 1) // TILE_SIZE): # Load A’s tile
    if row < n and (m*TILE_SIZE + tx) < n:
      tileA[ty, tx] = A[row, m*TILE_SIZE + tx]
    else:
      tileA[ty, tx] = 0

  # Load B’s tile
  if col < n and (m*TILE_SIZE + ty) < n:
    tileB[ty, tx] = B[m*TILE_SIZE + ty, col]
  else:
    tileB[ty, tx] = 0

  # Synchronize to make sure all threads loaded tiles
  cuda.syncthreads()

  # Multiply current tiles
  for k in range(TILE_SIZE):
    temp += tileA[ty, k] * tileB[k, tx]

  # Synchronize before loading new tiles
  cuda.syncthreads()

  #store result
  if row < n and col < n:
    C[row, col] = temp
```

**Host (Driver) code**

```python
N = 5
A = np.zeros((N, N), dtype=np.int32)
B = np.zeros((N, N), dtype=np.int32)
for i in range(N):
  for j in range(N):
    A[i, j] = i + j
    B[i, j] = i * j

#Print input matrices
print("Matrix A:\n", A)
print("Matrix B:\n", B)
```

```python
#Copy matrices to GPU
A_global = cuda.to_device(A)
B_global = cuda.to_device(B)
C_global = cuda.device_array((N, N), dtype=np.int32)
```

```python
#Configure CUDA grid G blocks
threads_per_block = (TILE_SIZE, TILE_SIZE)
blocks_per_grid_x = (N + TILE_SIZE - 1) // TILE_SIZE
blocks_per_grid_y = (N + TILE_SIZE - 1) // TILE_SIZE
blocks_per_grid = (blocks_per_grid_x, blocks_per_grid_y)

```

```python
#Launch kernel
tiled_matmul[blocks_per_grid, threads_per_block](A_global, B_global, C_global, N)
```

```python
#Copy result back
C = C_global.copy_to_host()
```

```python
#Print result
print("Product Matrix C:\n", C)
```
