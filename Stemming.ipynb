```python
#Text preprocessing using stemming
```

**Stemming** is a text preprocessing technique in NLP that reduces words to their root or base form (the "stem") by stripping away prefixes and suffixes.

Roots of a word ->known as Lemma

```python
words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

**PorterStemmer**

```python
from nltk.stem import PorterStemmer
stemming=PorterStemmer()
```

```python
for word in words:
  print(word+"---->",stemming.stem(word))
```

```python
stemming.stem("congratulations")
```

**RegexpStemmer Class**

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$', min=4)
#  Any substrings that match the regular expressions will be removed.
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('cars')
```

```python
reg_stemmer.stem('car')
```

```python
reg_stemmer.stem('relatable')
```

**Snowball Stemmer**

```python
from nltk.stem import SnowballStemmer
snowballstemmer=SnowballStemmer('english')
```

```python
for word in words:
  print(word+"------->",snowballstemmer.stem(word))
```

```python
stemming.stem('finally'),stemming.stem("succesfully")
```
