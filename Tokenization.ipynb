```python
corpus="""Hello, world! Welcome to the Python NLP Class.
Don't you think the world's changing faster than we'd expect?Dr. Smith spent $30,000,000 last year, seeing a profit increase of 40.75% . Machine learning is fascinating! Artificial Intelligence is an emerging technology. """
```

```python
print (corpus)
```

```python
#Tokenization
##sentence-->paragraphs
from nltk.tokenize import sent_tokenize
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
documents=sent_tokenize(corpus) #  . and ! create another sentence
```

```python
type(documents)
```

```python
for sentence in documents:
  print(sentence)
```

```python
##Tokenization
# paragraph-->words
# sentence-->words
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for sentence in documents:
  print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordDetokenizer
tokenizer=TreebankWordDetokenizer()
```

```python
tokenizer.tokenize(corpus)
```
