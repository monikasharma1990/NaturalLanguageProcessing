**Text Preprocessing Lemmatization**

**Wordnet Lemmatizer**

Lemmatization technique is like stemming. The output we will get after lemmatization is called 'lemma' which is a root word rather than a word stem , the output of Stemming. After Lemmatization, we will be getting a valid word that means the same thing

**Use cases :**
Q & A,
Chatbots,
text summarisation


```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
import nltk
nltk.download('wordnet')
```

```python
lemmatizer.lemmatize("eating",pos='v')
```

```python
words=['eat','eating','eaten','history','program','programming','finally','writing','writes','written','fairly']
```

```python
for word in words:
  print(word,"--->",lemmatizer.lemmatize(word,pos='v'))
```
