This file contains practice work for the different regular expressions used in 'speech and language processing'.

Regular expression search requires a **pattern** that we want to search
for, and a **corpus of texts** to search through.

**Anchors**:
1. \b  : word boundary
2. \B  : non-boundary
3. caret ^
4. Dollar sign $

```python
!echo "cat scatter category cat" | grep -P -o '\bcat'
```

```python
!echo "abc99 99abc 99 xyz99" | grep -P -o '\b99'
```

```python
!echo "abc99 99abc 99 xyz99" | grep -P -o '99\b'
```

```python
!echo "hello hello_world hello-world" | grep -P -o '\bhello\b'
```

```python
!echo "apple pineapple snapple apple" | grep -P -o '\bapple\b'
```

```python
!echo "cat concatenate bobcat cat" | grep -P -o '\Bcat\B'
```

```python
!echo "123 123abc abc123" | grep -P -o '\b123\b'
```

```python
!echo "car cart scar cartoon" | grep -P -o '\bcar'
```

```python
!echo "car cart scar cartoon" | grep -P -o 'car\b'
```

```python
!echo "cat_cat cat-cat (cat)" | grep -P -o '\bcat\b'
```

```python
!echo "scatter cat category" | grep -P -o '\Bcat\B'
```

```python
!echo "playing running sing" | grep -P -o '\Bing\b'
```

```python
!echo "cat is sleeping" | grep -P -o '^cat'
```

If the caret ˆ is the first symbol after
the open square brace [, the resulting pattern is negated. For example, the
pattern /[ˆa]/ matches any single character (including special characters)
except a

```python
# caret ^ as negation when used inside  []
!echo "cat is sleeping" | grep -P -o '[^cat]'
```

Q1: Write a regular expression to find strings like $3.88 ?

```python
!echo 'I need $3.88 for my pocket money' | grep -P -o '\$[0-9]+(\.[0-9]+)?'
```

Q2: Write a regular expression to find stings like 35% ?

```python
!echo 'This mixture contain 35% sugar' | grep -P -o '[0-9]+\%'
```

Quantifiers :

| Quantifier | Meaning               |
| ---------- | --------------------- |
| `*`        | 0 or more             |
| `+`        | 1 or more             |
| `?`        | 0 or 1                |
| `{n}`      | exactly n times       |
| `{n,}`     | at least n times      |
| `{n,m}`    | between n and m times |


```python
!echo 'color colour colouur' | grep -P -o 'colou*r'
```

```python
!echo 'color colour colouur' | grep -P -o 'colou+r'
```

```python
!echo 'color colour colouur' | grep -P -o 'colou?r'
```

```python
!echo 'color colour colouur' | grep -P -o 'colou{2}r'
```
