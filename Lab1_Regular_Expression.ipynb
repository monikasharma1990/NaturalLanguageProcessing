`grep -E '^A[0-9]+\sThere' TMIB_SAMPLE.txt`


The above expression will account for line numbers (common in corpus files), space and also check for the word we want to search for. Note the use of the character class for the line numbers, as well as Kleene +. Grep uses the notation `\s` to indicate space. >
>`\b` Match the empty string at the edge of a word.

>`\<` Match the empty string at the beginning of word.  
>`\>` Match the empty string at the end of word.  
>`\w` Match word constituent, it is a synonym for ‘[_[:alnum:]]’.  
>`\W` Match non-word constituent, it is a synonym for ‘[^_[:alnum:]]’.   
>`\s` Match whitespace, it is a synonym for ‘[[:space:]]’.  
>`\S` Match non-whitespace, it is a synonym for ‘[^[:space:]]’.

## A4

We can use the `\w` to search for specific patterns in the text. For instance, we may want to look for the subject NPs for the verb *be*.

`grep -E '\w+ (were|was|is|are)' TMIB_SAMPLE.txt`

Note that this will match both NPs and complementizers like *that* or wh-words like *which*. Then we may have to make the expression such that it *doesn't* match a word that precedes it. For this, we need to use a *negative lookbehind*. This simply says that 'what immediately precedes this string is not a which. For more complex expressions like lookbehinds and lookaheads, we need to use the -P flag which simply means, use Perl style regular expressions.


`grep -P '(?<!which) were' TMIB_SAMPLE.txt`

We can chain the two grep commands to get rid of the offending *that* or *which* cases

`grep -E '\w+ (were|was|is|are)' TMIB_SAMPLE.txt | grep -P '(?<!which) were' TMIB_SAMPLE.txt`

##### Character classes

Character classes are their own language, and we can use them to search for all instances of 'we' in this mini corpus, for example. This takes care of both upper and lower case *w*.

` grep -E '[Ww]e\>' TMIB_SAMPLE.txt `

However, note that there's a difference between the *or* metacharacter and character class. For instance, the same expression above may be written as:-

`grep -E '(W|w)e\>' TMIB_SAMPLE.txt `

But not as

`grep -E '(W|we)\>' TMIB_SAMPLE.txt`

Why?

Character classes might also be used to negate, i.e. display what is not listed. E.g. [^1-6] matches a character that is not 1-6. Perhaps we might want to match everything in the corpus except the line numbers. Then we can try the following:

`grep -E [^A0-9+] TMIB_SAMPLE.txt `

This will match everything but the line numbers. Note that the caret *inside* the character class brackets has a totally different meaning! It doesn't mean 'start of line' anymore.

##### Wildcards

The metacharacter `.` is a shorthand for a character class that will match any character. While searching for dates in the following formats

>6.7.1998  
06-07-1998  
6/07/1998  

` grep -E '[0-9]+.[0-9]+.[0-9]+' word_list.txt `

The above regex will match all of these cases, as the wildcard represents 'any character'. However, note that if someone makes a mistake in the file, the wildcard will match anything, even a sequence like *19 20734 7639*.

It is safer to then use character classes. Note that once the wildcard metacharacter `.` is inside the character class bracket, it loses its wildcard meaning, and just becomes a period symbol:-

`grep -E '[0-9]+[./-][0-9]+[./-][0-9]+' word_list.txt `

Now, this will match the date format alone. Sometimes, it is quicker to use wildcards, but only if you are sure that the document does not contain possible other matches.


```python
!grep -E '[0-9]+.[0-9]+.[0-9]+' word_list.txt
```

## Regular expressions  for tokenization


One of the most important pre-processing tasks for NLP is tokenization. This simply means segmenting words in running text. This is actually not as simple as it may seem.



E.g. words contain the punctuations, hyphens, and some other words are kept as is e.g. *liver-pill*. What is the proper way to segment these forms?

![image](probs.png)

We will use some regular expression based tokenizers to split the text further.


```python
import nltk
from nltk.tokenize import RegexpTokenizer


tmib = '''There were four of us-George, and William Samuel Harris, and myself, and Montmorency.
We were sitting in my room, smoking, and talking about how bad we were-bad from a medical point of view I mean, of course.
We were all feeling seedy, and we were getting quite nervous about it.
Harris said he felt such extraordinary fits of giddiness come over him at times, that he hardly knew what he was doing; and then George said that he had fits of giddiness too, and hardly knew what he was doing.
With me, it was my liver that was out of order.
I knew it was my liver that was out of order, because I had just been reading a patent liver-pill circular, in which were detailed the various symptoms by which a man could tell when his liver was out of order.
I had them all.'''

tokenizer = RegexpTokenizer('[-,;:.!?$\"]|\w+')
tokenizer.tokenize(tmib)

```

```python
# [-,;:.!?$\"]|\w+
#Now let's run our tokenizer on this sentence
sentence = 'Good muffins cost $3.88 in New York.  Please buy me two of them. Thanks.'
tokenizer.tokenize(sentence)
#it turns out that this regex is not that good at recognizing numeral amounts.
```

```python
sentence = 'Good muffins cost $3.88 in New York.  Please buy me two of them. Thanks.'

tokenizer2 = RegexpTokenizer('\w+|$[\d\.]+|\S+')
tokenizer2.tokenize(sentence)

#How can we modify this to include numerals like this: 88.5% or 50% or 1% ? Write a tokenizer3 regular expression
#below:-

#Q1: Write a regular expression to find strings like $3.88 ?
#Q2: Write a regular expression to find stings like 35% ?
```

```python
sentence = 'Good muffins cost $3.88 and 88% in New York.  Please buy me two of them. Thanks.'

tokenizer2 = RegexpTokenizer('\w+|$[\d\.]+|\S+')
tokenizer2.tokenize(sentence)

#How can we modify this to include numerals like this: 88.5% or 50% or 1% ? Write a tokenizer3 regular expression
#below:-
# \d+(?:\.\d+)?%
```

>
> `*` zero or more occurrences of the previous char or expression.
> `+` one or more occurrences of the previous char or expression.
> `?` exactly zero or one occurrence of the previous char or expression.
> `{n}` n occurrences of the previous char or expression.
> `{n,m}` from n to m occurrences of the previous char or expression.
> `{n,}` at least n occurrences of the previous char or expression.
> `{,m}` up to m occurrences of the previous char or expression.
